```python
df.dropna(inplace = True)
df.loc[:, "location":"val"] # выбираем все строки, а столбцы от "location" до "val"
```

```python
df.nlargest(60,'val') # 99 наибольших значений по столбцу 'val'
```

```python
df.groupby('cause').agg('mean') # среднее значение по всем числовым столбцам для каждой причины смерти
```

```python
df_cause = df.groupby('cause').agg({'year': ['count'],
                        'val': ['mean', 'median']})
# статистика по причинам смерти 
# для столбца year считаем count (число уникальных значений), для 'val' — среднее и медиану
df_cause
```
